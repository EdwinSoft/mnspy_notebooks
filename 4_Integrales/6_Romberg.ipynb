# Integración de Romberg
La **integración de Romberg** es un método de alta precisión para la integración numérica. Se basa en la **regla del trapecio** y utiliza una técnica llamada **extrapolación de Richardson** para mejorar sistemáticamente las estimaciones.

El proceso comienza calculando la integral con la regla del trapecio para diferentes números de segmentos (1, 2, 4, 8, ...). Luego, estas estimaciones se combinan de forma iterativa para cancelar los términos de error de orden inferior, lo que conduce a una convergencia mucho más rápida hacia el valor verdadero de la integral.

El resultado es una tabla triangular donde las estimaciones en la diagonal son las más precisas. Debido a su eficiencia y precisión, es un método muy potente.

## Ejemplo

**Problema:**

Utilice la integración de Romberg para estimar la integral de la siguiente función desde $x=0$ hasta $x=0.8$:
$$ f(x) = 0.2 + 25x - 200x^2 + 675x^3 - 900x^4 + 400x^5 $$

```python
from mnspy import Romberg
```

Primero, definimos la función a integrar.

```python
def f(x):
    return 0.2 + 25 * x - 200 * x ** 2 + 675 * x ** 3 - 900 * x ** 4 + 400 * x ** 5
```

- Creamos una instancia de la clase `Romberg`.
- Le pasamos la función `f`, el límite inferior `a=0` y el límite superior `b=0.8`, por defecto trabaja con una `tolerancia=1e-8`.
- La clase realiza internamente las iteraciones de la extrapolación de Richardson.

```python
ro = Romberg(f, 0, 0.8)
print(f"El valor aproximado de la integral es: {ro.integral}")
```

La gráfica muestra la función a integrar en el intervalo especificado.

```python
ro.graficar()
```

### Verificación
 
El valor analítico de la integral es el mismo que en los ejemplos de Simpson:
$$ \int_{0}^{0.8} f(x) dx = 1.640533... $$

El resultado obtenido con el método de Romberg es extremadamente preciso, coincidiendo con el valor exacto hasta muchas cifras decimales, lo que demuestra la potencia del método.
