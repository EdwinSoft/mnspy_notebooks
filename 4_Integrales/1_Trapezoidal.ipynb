# Método del Trapecio
El **método del trapecio** es una técnica de integración numérica que aproxima el valor de una integral definida. La idea es dividir el área bajo la curva de la función en varios trapecios y sumar sus áreas.

Para un solo trapecio entre los puntos $a$ y $b$, el área es:
$$ I \approx (b-a) \frac{f(a) + f(b)}{2} $$

Para mejorar la precisión, se puede aplicar la **regla del trapecio múltiple**, dividiendo el intervalo $[a, b]$ en $n$ segmentos de igual ancho $h = (b-a)/n$. La fórmula es:
$$ I \approx \frac{h}{2} \left[ f(x_0) + 2\sum_{i=1}^{n-1}f(x_i) + f(x_n) \right] $$

## Ejemplo
 
 **Problema:**
 
 Utilice la regla del trapecio con 12 segmentos para estimar la integral de la siguiente función desde $x=1$ hasta $x=2$:
 $$ f(x) = \left(x + \frac{1}{x}\right)^2 $$

```python
from mnspy import Trapezoidal
```

```python
# Primero, definimos la función a integrar.
def f(x):
    return (x + 1 / x) ** 2
```

- Creamos una instancia de la clase `Trapezoidal`.
- Le pasamos la función `f`, el límite inferior `a=1`, el límite superior `b=2` y el número de segmentos `n=12`.
- La clase calcula automáticamente el valor de la integral.

```python
trap = Trapezoidal(f, 1, 2, 12)
print(f"El valor aproximado de la integral es: {trap.integral}")
```

```python
# La gráfica muestra la función original y los 12 trapecios utilizados para aproximar el área bajo la curva.
trap.graficar()
```

### Verificación
 
El valor analítico de la integral es:
$$ \int_1^2 \left(x + \frac{1}{x}\right)^2 dx = \int_1^2 \left(x^2 + 2 + x^{-2}\right) dx = \left[ \frac{x^3}{3} + 2x - \frac{1}{x} \right]_1^2 $$
$$ = \left( \frac{8}{3} + 4 - \frac{1}{2} \right) - \left( \frac{1}{3} + 2 - 1 \right) = \frac{29}{6} \approx 4.8333... $$

El resultado numérico obtenido es muy cercano al valor exacto.
