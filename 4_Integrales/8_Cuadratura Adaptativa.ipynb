# Cuadratura Adaptativa
La **cuadratura adaptativa** es una estrategia de integración numérica que ajusta dinámicamente el tamaño de los subintervalos para lograr una precisión deseada con la menor cantidad de cálculos posible.

A diferencia de los métodos que usan un tamaño de paso fijo (como la regla del trapecio o Simpson con un `n` predefinido), la cuadratura adaptativa:
1.  Estima la integral y el error en un subintervalo.
2.  Si el error es mayor que la tolerancia especificada, el subintervalo se divide en dos y el proceso se aplica recursivamente a cada nueva mitad.
3.  Si el error es aceptable, se suma el valor de la integral de ese subintervalo al total y no se refina más.

Esto permite que el algoritmo concentre el esfuerzo computacional en las regiones donde la función es más "difícil" de integrar (es decir, donde cambia rápidamente), mientras que usa intervalos más grandes en las regiones donde la función es suave.


## Ejemplo
 
**Problema:**

Utilice la cuadratura adaptativa para estimar la integral de la siguiente función desde $x=0$ hasta $x=0.8$, con una tolerancia de `0.001`.
$$ f(x) = 0.2 + 25x - 200x^2 + 675x^3 - 900x^4 + 400x^5 $$

```python
from mnspy import CuadraturaAdaptativa
%matplotlib inline
```

Primero, definimos la función a integrar.

```python
def f(x):
    return 0.2 + 25 * x - 200 * x ** 2 + 675 * x ** 3 - 900 * x ** 4 + 400 * x ** 5
```

- Creamos una instancia de la clase `CuadraturaAdaptativa`.
- Le pasamos la función, el intervalo [0, 0.8] y la tolerancia deseada.
- El algoritmo refinará los subintervalos hasta que el error estimado sea menor que `tol`.

```python
ca = CuadraturaAdaptativa(f, 0, 0.8, tol=0.001)
print(f"El valor aproximado de la integral es: {ca.integral}")
```

La gráfica muestra la función a integrar.

```python
ca.graficar()
```

### Verificación
 
El valor analítico de la integral es el mismo que en los ejemplos anteriores:
$$ \int_{0}^{0.8} f(x) dx = 1.640533... $$

El resultado obtenido con la cuadratura adaptativa es muy preciso, cumpliendo con la tolerancia especificada.
