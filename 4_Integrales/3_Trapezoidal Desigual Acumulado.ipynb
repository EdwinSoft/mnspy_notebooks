 # Método Trapezoidal Acumulado para Segmentos Desiguales
El cálculo de la **integral acumulada** consiste en determinar el valor de la integral desde el punto de inicio hasta cada punto subsiguiente del dominio. Es decir, se calcula $I(x) = \int_{a}^{x} f(t) dt$ para varios valores de $x$.

Cuando se trabaja con datos tabulados (potencialmente con espaciado desigual), este método aplica la regla del trapecio de forma sucesiva. El resultado no es un único valor, sino un vector que contiene la integral acumulada en cada punto.

La integral acumulada en el punto $x_k$ se calcula como:
$$ I(x_k) = \sum_{i=0}^{k-1} (x_{i+1} - x_i) \frac{y_i + y_{i+1}}{2} $$

Esto es útil para visualizar cómo se acumula una cantidad a lo largo del tiempo o del espacio, como la distancia recorrida a partir de datos de velocidad.


## Ejemplo
 
**Problema:**

Calcular y visualizar la integral acumulada de la función $f(x) = \cos(x)$ en el intervalo $[0, 2\pi]$, utilizando 40 puntos de datos.


```python
from mnspy import TrapezoidalDesigualAcumulado
import numpy as np
from tabulate import tabulate
```

```python
# Definimos los vectores `x` e `y` con los datos de la función.
# Aunque `linspace` crea puntos espaciados uniformemente, el método funciona igual para datos desiguales.
x = np.linspace(0, 2 * np.pi, 40)
y = np.cos(x)
```

```python
# Creamos una instancia de la clase `TrapezoidalDesigualAcumulado`.
# La clase calcula automáticamente el vector de la integral acumulada.
trap = TrapezoidalDesigualAcumulado(x, y)
# El método `graficar()` muestra dos subtramas:
# 1. La función original (y vs x).
# 2. La integral acumulada calculada numéricamente (I vs x).
trap.graficar()
```

```python
# Mostramos los datos originales y el resultado de la integral acumulada en una tabla.
# El vector `trap.integral_acumulada` contiene el resultado.
tabla_datos = {'x': x, 'y = cos(x)': y, 'Integral Acumulada': trap.integral}
tabulate(tabla_datos, tablefmt='html', headers='keys', floatfmt=".4f")
```

### Verificación
 
El valor analítico de la integral acumulada de $\cos(x)$ desde 0 es:
$$ \int_0^x \cos(t) dt = [\sin(t)]_0^x = \sin(x) - \sin(0) = \sin(x) $$

Podemos comparar nuestro resultado numérico con la función $\sin(x)$ para verificar la precisión. La gráfica de la integral acumulada obtenida numéricamente debe ser muy similar a la gráfica de la función seno, como se observa en la segunda subtrama de la figura anterior.
