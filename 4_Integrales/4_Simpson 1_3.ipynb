# Regla de Simpson 1/3
La **regla de Simpson 1/3** es un método de integración numérica que mejora la precisión del método del trapecio al aproximar la función a integrar no con líneas rectas, sino con **parábolas (polinomios de segundo grado)**.

Para aplicar la regla, se toman tres puntos consecutivos y se hace pasar una parábola por ellos. La fórmula para una sola aplicación sobre un intervalo con puntos $x_0, x_1, x_2$ es:
$$ I \approx \frac{h}{3} [f(x_0) + 4f(x_1) + f(x_2)] $$

Para mejorar la precisión, se puede aplicar la **regla de Simpson 1/3 múltiple**, que requiere que el intervalo de integración se divida en un **número par de segmentos ($n$)**. La fórmula general es:
$$ I \approx \frac{h}{3} \left[ f(x_0) + 4\sum_{i=1,3,5}^{n-1}f(x_i) + 2\sum_{j=2,4,6}^{n-2}f(x_j) + f(x_n) \right] $$

Este método es generalmente mucho más preciso que la regla del trapecio.


## Ejemplo
 
**Problema:**

Utilice la regla de Simpson 1/3 para estimar la integral de la siguiente función desde $x=0$ hasta $x=0.8$:
$$ f(x) = 0.2 + 25x - 200x^2 + 675x^3 - 900x^4 + 400x^5 $$

```python
from mnspy import Simpson13
```

Primero, definimos la función a integrar.

```python
def f(x):
    return 0.2 + 25 * x - 200 * x ** 2 + 675 * x ** 3 - 900 * x ** 4 + 400 * x ** 5
```

### Caso 1: 2 Segmentos
Creamos una instancia de la clase `Simpson13` con n=2 segmentos.

```python
sim_2 = Simpson13(f, 0, 0.8, 2)
print(f"El valor aproximado de la integral con 2 segmentos es: {sim_2.integral}")
```

La gráfica muestra la función original y la parábola utilizada para la aproximación.

```python
sim_2.graficar()
```

### Caso 2: 4 Segmentos
Ahora, repetimos el proceso con n=4 segmentos para mejorar la precisión.

```python
sim_4 = Simpson13(f, 0, 0.8, 4)
print(f"El valor aproximado de la integral con 4 segmentos es: {sim_4.integral}")
```

La gráfica ahora muestra las dos parábolas (una de x=0 a x=0.4 y otra de x=0.4 a x=0.8) que se usan para la aproximación.

```python
sim_4.graficar()
```
