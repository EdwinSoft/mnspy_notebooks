# Método del Trapecio para Segmentos Desiguales
En muchas aplicaciones prácticas, especialmente al trabajar con datos experimentales, no se tiene la garantía de que los puntos de datos estén espaciados uniformemente. En estos casos, la fórmula de la regla del trapecio múltiple no se puede aplicar directamente.

Sin embargo, la lógica del método del trapecio sigue siendo válida. La solución es calcular el área de cada trapecio individual (formado por cada par de puntos consecutivos) y luego sumar todas estas áreas.

El área de un trapecio individual entre los puntos $(x_i, y_i)$ y $(x_{i+1}, y_{i+1})$ es:
$$ I_i = (x_{i+1} - x_i) \frac{y_i + y_{i+1}}{2} $$

La integral total es la suma de las áreas de todos los trapecios:
$$ I = \sum_{i=0}^{n-1} I_i $$


## Ejemplo
 
**Problema:**

Estimar la integral del siguiente conjunto de datos tabulados, que representan mediciones tomadas a intervalos de tiempo desiguales.


```python
from mnspy import TrapezoidalDesigual
import numpy as np
from tabulate import tabulate
# %matplotlib widget
```

```python
# Definimos los vectores `x` (variable independiente) e `y` (variable dependiente) con los datos tabulados.
x = np.array([0, 0.12, 0.22, 0.32, 0.36, 0.4, 0.44, 0.54, 0.64, 0.7, 0.8])
y = np.array([0.2, 1.309729, 1.305241, 1.743393, 2.074903, 2.456, 2.842985, 3.507297, 3.181929, 2.363, 0.232])
```

```python
# Usamos la biblioteca `tabulate` para mostrar los datos de forma clara y ordenada.
tabulate({'x': x, 'y': y},headers='keys',tablefmt='html')
```

```python
# Creamos una instancia de la clase `TrapezoidalDesigual`, pasándole los vectores de datos.
# La clase calcula automáticamente la integral sumando las áreas de los trapecios individuales.
trap = TrapezoidalDesigual(x, y)
print(f"El valor aproximado de la integral es: {trap.integral}")
```

```python
trap.graficar()
```
