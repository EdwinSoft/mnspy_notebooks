# Cuadratura de Gauss-Legendre
La **cuadratura de Gauss-Legendre** es una de las técnicas de integración numérica más potentes y precisas. A diferencia de los métodos de Newton-Cotes (Trapecio, Simpson), que utilizan puntos de evaluación espaciados uniformemente, la cuadratura de Gauss elige las ubicaciones de los puntos de forma óptima.

La idea es aproximar la integral como una suma ponderada:
$$ I \approx \sum_{i=1}^{n} c_i f(x_i) $$

Tanto los **pesos ($c_i$)** como los **puntos de evaluación ($x_i$)** se eligen específicamente para maximizar la precisión. Estos valores se derivan de las raíces de los polinomios de Legendre.

La fórmula estándar está definida para el intervalo $[-1, 1]$. Para aplicar el método a un intervalo arbitrario $[a, b]$, se realiza un cambio de variable.

La principal ventaja es que una fórmula de Gauss-Legendre de $n$ puntos puede integrar exactamente un polinomio de grado hasta $2n-1$, lo que la hace mucho más precisa que los métodos de Newton-Cotes para el mismo número de evaluaciones de la función.


## Ejemplo
 
**Problema:**

Utilice la cuadratura de Gauss-Legendre con 6 puntos para estimar la integral de la siguiente función desde $x=0$ hasta $x=0.8$:
$$ f(x) = 0.2 + 25x - 200x^2 + 675x^3 - 900x^4 + 400x^5 $$

```python
from mnspy import GaussLegendre
%matplotlib inline
```

Primero, definimos la función a integrar.

```python
def f(x):
    return 0.2 + 25 * x - 200 * x ** 2 + 675 * x ** 3 - 900 * x ** 4 + 400 * x ** 5
```

- Creamos una instancia de la clase `GaussLegendre`.
- Le pasamos la función `f`, el intervalo `[0, 0.8]` y el número de puntos `n=6`.
- La clase realiza el cambio de variable y aplica la fórmula de cuadratura internamente.

```python
gl = GaussLegendre(f, 0, 0.8, 6)
print(f"El valor aproximado de la integral es: {gl.integral}")
```

La gráfica muestra la función a integrar en el intervalo especificado.

```python
gl.graficar()
```

### Verificación
 
El valor analítico de la integral es el mismo que en los ejemplos anteriores:
$$ \int_{0}^{0.8} f(x) dx = 1.64053333... $$

El resultado obtenido con Gauss-Legendre de 6 puntos es **exacto** al valor analítico. Esto se debe a que la fórmula de 6 puntos puede integrar exactamente polinomios de grado hasta $2 \times 6 - 1 = 11$. Como nuestra función es un polinomio de grado 5, el método proporciona el resultado exacto (salvo pequeños errores de punto flotante).
