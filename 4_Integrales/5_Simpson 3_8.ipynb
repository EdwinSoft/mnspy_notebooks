# Regla de Simpson 3/8
La **regla de Simpson 3/8** es otro método de integración numérica que, al igual que la regla 1/3, se basa en ajustar polinomios a los datos. En este caso, se utilizan **polinomios cúbicos (de tercer grado)** que pasan por cuatro puntos consecutivos.

La fórmula para una sola aplicación sobre un intervalo con puntos $x_0, x_1, x_2, x_3$ es:
$$ I \approx \frac{3h}{8} [f(x_0) + 3f(x_1) + 3f(x_2) + f(x_3)] $$

Para aplicar la regla de forma múltiple, el intervalo de integración debe dividirse en un **número de segmentos ($n$) que sea múltiplo de 3**.

Aunque es un poco más precisa que la regla 1/3, su uso es menos común debido a la restricción más estricta sobre el número de segmentos.


## Ejemplo
 
**Problema:**

Utilice la regla de Simpson 3/8 para estimar la integral de la siguiente función desde $x=0$ hasta $x=0.8$:
$$ f(x) = 0.2 + 25x - 200x^2 + 675x^3 - 900x^4 + 400x^5 $$

```python
from mnspy import Simpson38
```

Primero, definimos la función a integrar.

```python
def f(x):
    return 0.2 + 25 * x - 200 * x ** 2 + 675 * x ** 3 - 900 * x ** 4 + 400 * x ** 5
```

### Caso 1: 3 Segmentos
Creamos una instancia de la clase `Simpson38` con n=3 segmentos.

```python
sim_3 = Simpson38(f, 0, 0.8, 3)
print(f"El valor aproximado de la integral con 3 segmentos es: {sim_3.integral}")
```

La gráfica muestra la función original y el polinomio cúbico utilizado para la aproximación.

```python
sim_3.graficar()
```

### Caso 2: 6 Segmentos
Ahora, repetimos el proceso con n=6 segmentos para mejorar la precisión.

```python
sim_6 = Simpson38(f, 0, 0.8, 6)
print(f"El valor aproximado de la integral con 6 segmentos es: {sim_6.integral}")
```

La gráfica ahora muestra los dos polinomios cúbicos que se usan para la aproximación.

```python
sim_6.graficar()
```

### Verificación
 
El valor analítico de la integral es el mismo que en el ejemplo de Simpson 1/3:
$$ \int_{0}^{0.8} f(x) dx = 1.640533... $$

Comparando los resultados:
- **Valor exacto:** 1.640533
- **Simpson 3/8 (n=6):** 1.638632
- **Simpson 3/8 (n=3):** 1.519170

Al igual que con otros métodos, aumentar el número de segmentos (manteniéndolo como múltiplo de 3) mejora significativamente la precisión de la aproximación.
