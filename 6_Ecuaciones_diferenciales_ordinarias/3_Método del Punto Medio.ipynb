# Método del Punto Medio (Runge-Kutta de 2º Orden)
El **método del punto medio** es otra mejora del método de Euler y, al igual que el método de Heun, es un método de Runge-Kutta de segundo orden.

La estrategia aquí es:
1.  Usar el método de Euler para estimar el valor de $y$ en el **punto medio** del intervalo ($x_i + h/2$).
2.  Utilizar la pendiente calculada en ese punto medio como una aproximación de la pendiente promedio para todo el intervalo.
3.  Calcular el valor final de $y_{i+1}$ usando la pendiente del punto medio.

Las fórmulas son:
$$ y_{i+1/2} = y_i + f(x_i, y_i) \frac{h}{2} $$
$$ y_{i+1} = y_i + f(x_i + \frac{h}{2}, y_{i+1/2}) h $$

Este método tiene un error de truncamiento local de orden $O(h^2)$, lo que lo hace mucho más preciso que el método de Euler.


## Ejemplo
 
**Problema:**

Resolver la misma EDO de los ejemplos anteriores para comparar la precisión. El problema es resolver desde $x=0$ hasta $x=4$ con un tamaño de paso $h=0.1$:
$$ \frac{dy}{dx} = -2x^3 + 12x^2 - 20x + 8.5 $$
La condición inicial es $y(0) = 1$. La solución analítica exacta es:
$$ y(x) = -0.5x^4 + 4x^3 - 10x^2 + 8.5x + 1 $$

```python
from mnspy import PuntoMedio
```

# --- 1. Definición de las Funciones ---

```python
# Definimos la función f(x, y) que representa la EDO dy/dx.
def f(x, y):
    return -2 * x ** 3 + 12 * x ** 2 - 20 * x + 8.5

# Definimos la solución analítica exacta para comparar los resultados.
def sol_exacta(x):
    return -0.5 * x ** 4 + 4 * x ** 3 - 10 * x ** 2 + 8.5 * x + 1
```

# --- 2. Solución Numérica ---

```python
# Creamos una instancia de la clase `PuntoMedio`.
# Los argumentos son los mismos que para las clases anteriores.
# (función_edo, x_inicial, x_final, y_inicial, tamaño_paso, solucion_exacta)
pm = PuntoMedio(f, 0, 4, 1, 0.1, sol_exacta)
# El método `graficar()` muestra una comparación visual.
# Al igual que con el método de Heun, la solución del punto medio se ajusta casi perfectamente a la solución exacta.
pm.graficar()
```

# --- 3. Tabla de Resultados ---

```python
# El método `solucion()` muestra la tabla de resultados.
# El error en cada paso es muy pequeño, demostrando la mejora significativa sobre el método de Euler.
pm.solucion()
```
