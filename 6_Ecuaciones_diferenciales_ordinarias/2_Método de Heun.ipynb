# Método de Heun (Predictor-Corrector)
El **método de Heun** es una mejora del método de Euler. Es un método de **predictor-corrector** que utiliza la pendiente en el punto inicial y una estimación de la pendiente en el punto final para obtener una mejor aproximación del promedio de la pendiente en el intervalo.

1.  **Paso Predictor:** Se estima un valor preliminar de $y_{i+1}$ usando el método de Euler.
 $$ y_{i+1}^0 = y_i + f(x_i, y_i) h $$

2.  **Paso Corrector:** Se utiliza el valor predicho para calcular una pendiente corregida y el valor final de $y_{i+1}$.
 $$ y_{i+1} = y_i + \frac{f(x_i, y_i) + f(x_{i+1}, y_{i+1}^0)}{2} h $$

Este método es un ejemplo de un método de Runge-Kutta de segundo orden y tiene un error de truncamiento local de orden $O(h^2)$, lo que lo hace significativamente más preciso que el método de Euler.


## Ejemplo
 
**Problema:**

Resolver la misma EDO del ejemplo de Euler para comparar la precisión. El problema es resolver desde $x=0$ hasta $x=4$ con un tamaño de paso $h=0.1$:
$$ \frac{dy}{dx} = -2x^3 + 12x^2 - 20x + 8.5 $$
La condición inicial es $y(0) = 1$. La solución analítica exacta es:
$$ y(x) = -0.5x^4 + 4x^3 - 10x^2 + 8.5x + 1 $$

```python
from mnspy import Heun
```

# --- 1. Definición de las Funciones ---

```python
# Definimos la función f(x, y) que representa la EDO dy/dx.
def f(x, y):
    return -2 * x ** 3 + 12 * x ** 2 - 20 * x + 8.5

# Definimos la solución analítica exacta para comparar los resultados.
def sol_exacta(x):
    return -0.5 * x ** 4 + 4 * x ** 3 - 10 * x ** 2 + 8.5 * x + 1
```

# --- 2. Solución Numérica ---

```python
# Creamos una instancia de la clase `Heun`.
# Los argumentos son los mismos que para la clase Euler.
# (función_edo, x_inicial, x_final, y_inicial, tamaño_paso, solucion_exacta)
he = Heun(f, 0, 4, 1, 0.1, sol_exacta)
# El método `graficar()` muestra una comparación visual.
# Se puede observar que la solución de Heun (puntos) se ajusta casi perfectamente a la solución exacta (línea).
he.graficar()
```

# --- 3. Tabla de Resultados ---

```python
# El método `solucion()` muestra la tabla de resultados.
# Al comparar con el método de Euler, se observa que el error en cada paso es mucho menor.
he.solucion()
```
