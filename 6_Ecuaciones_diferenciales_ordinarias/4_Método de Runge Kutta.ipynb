# Métodos de Runge-Kutta (RK)
Los **métodos de Runge-Kutta (RK)** son una familia de métodos numéricos iterativos para la aproximación de soluciones de Ecuaciones Diferenciales Ordinarias (EDOs). Son los más utilizados en la práctica por su excelente balance entre precisión y eficiencia computacional.

La idea general es obtener una mayor precisión calculando la pendiente en varios puntos dentro del intervalo y promediándolas de forma ponderada. La forma general es:
$$ y_{i+1} = y_i + \phi(x_i, y_i, h)h $$
donde $\phi$ es el **incremento de la función**, que representa una pendiente promedio ponderada sobre el intervalo.

El método RK más famoso es el **RK de 4º orden**, pero existen métodos de diferentes órdenes. Los métodos de Heun y del Punto Medio son, de hecho, métodos RK de 2º orden.


## Ejemplo

**Problema:**

Resolver la misma EDO de los ejemplos anteriores para comparar la precisión. El problema es resolver desde $x=0$ hasta $x=4$ con un tamaño de paso $h=0.5$:
$$ \frac{dy}{dx} = -2x^3 + 12x^2 - 20x + 8.5 $$
La condición inicial es $y(0) = 1$. La solución analítica exacta es:
$$ y(x) = -0.5x^4 + 4x^3 - 10x^2 + 8.5x + 1 $$

```python
from mnspy import RungeKutta
```

# --- 1. Definición de las Funciones ---

```python
# Definimos la función f(x, y) que representa la EDO dy/dx.
def f(x, y):
    return -2 * x ** 3 + 12 * x ** 2 - 20 * x + 8.5

# Definimos la solución analítica exacta para comparar los resultados.
def sol_exacta(x):
    return -0.5 * x ** 4 + 4 * x ** 3 - 10 * x ** 2 + 8.5 * x + 1
```

# --- 2. Solución con Métodos RK de 2º Orden ---
Los métodos de Heun, Punto Medio y Ralston son todos de 2º orden. Se diferencian en cómo ponderan las pendientes.


```python
print("--- Método RK de 2º Orden: Punto Medio ---")
rk_pm = RungeKutta(f, 0, 4, 1, 0.5, a_2='punto_medio', sol_exacta=sol_exacta)
rk_pm.graficar()
```

```python
rk_pm.solucion()
```

```python
print("\n--- Método RK de 2º Orden: Heun ---")
rk_heun = RungeKutta(f, 0, 4, 1, 0.5, a_2='heun', sol_exacta=sol_exacta)
rk_heun.graficar()
```

```python
rk_heun.solucion()
```

```python
print("\n--- Método RK de 2º Orden: Ralston ---")
rk_ralston = RungeKutta(f, 0, 4, 1, 0.5, a_2='ralston', sol_exacta=sol_exacta)
rk_ralston.graficar()
```

```python
rk_ralston.solucion()
```

# --- 3. Solución con Método RK de Orden Superior (5º Orden) ---
Ahora, usamos un método de orden superior. Se espera que sea mucho más preciso que los de 2º orden para el mismo tamaño de paso.

```python
print("\n--- Método RK de 5º Orden ---")
rk5 = RungeKutta(f, 0, 4, 1, 0.5, orden=5, sol_exacta=sol_exacta)
rk5.graficar()
```

```python
rk5.solucion()
```

### Conclusión

Al observar las tablas de resultados, se puede notar una diferencia drástica en la precisión:

- **Métodos RK de 2º Orden (h=0.5):** Acumulan un error considerable al final del intervalo (el error final es de aproximadamente 1.3).
- **Método RK de 5º Orden (h=0.5):** La solución numérica es prácticamente idéntica a la solución exacta. El error final es del orden de $10^{-5}$, demostrando la increíble mejora en la precisión que ofrecen los métodos de orden superior.

Por esta razón, métodos como el RK de 4º orden (o superiores) son el estándar en la mayoría de las aplicaciones científicas y de ingeniería.
