# Método de Euler para Ecuaciones Diferenciales Ordinarias (EDOs)
El **método de Euler** es el enfoque numérico más simple para resolver problemas de valor inicial para Ecuaciones Diferenciales Ordinarias (EDOs). La idea es utilizar la pendiente en el punto actual para extrapolar y encontrar el siguiente punto.

Dada una EDO de la forma:
$$ \frac{dy}{dx} = f(x, y) $$
Con una condición inicial $y(x_0) = y_0$, la fórmula de Euler para encontrar el siguiente valor $y_{i+1}$ es:
$$ y_{i+1} = y_i + f(x_i, y_i) h $$
donde $h$ es el tamaño del paso.

Aunque es fácil de implementar, su precisión es limitada (error de orden $O(h)$), y errores pequeños pueden acumularse a lo largo de la integración.

## Ejemplo

**Problema:**

Resolver la siguiente EDO desde $x=0$ hasta $x=4$ con un tamaño de paso $h=0.1$:
$$ \frac{dy}{dx} = -2x^3 + 12x^2 - 20x + 8.5 $$
La condición inicial es $y(0) = 1$. La solución analítica exacta es:
$$ y(x) = -0.5x^4 + 4x^3 - 10x^2 + 8.5x + 1 $$

```python
from mnspy import Euler
```

# --- 1. Definición de las Funciones ---

```python
# Definimos la función f(x, y) que representa la EDO dy/dx.
def f(x, y):
    return -2 * x ** 3 + 12 * x ** 2 - 20 * x + 8.5
# Definimos la solución analítica exacta para comparar los resultados.
def sol_exacta(x):
    return -0.5 * x ** 4 + 4 * x ** 3 - 10 * x ** 2 + 8.5 * x + 1
```

# --- 2. Solución Numérica ---

```python
# Creamos una instancia de la clase `Euler`.
# Argumentos: (función_edo, x_inicial, x_final, y_inicial, tamaño_paso, solucion_exacta)
# En este caso: (f, 0, 4, 1, 0.1, sol_exacta)
eu = Euler(f, 0, 4, 1, 0.1, sol_exacta)
# El método `graficar()` muestra una comparación visual entre la solución numérica (Euler) y la solución exacta.
# Se puede observar cómo la solución de Euler se desvía de la curva real a medida que avanza.
eu.graficar()
```

# --- 3. Tabla de Resultados ---

```python
# El método `solucion()` muestra una tabla con los valores de x, y (numérico), y (exacto) y el error en cada paso.
eu.solucion()
```
