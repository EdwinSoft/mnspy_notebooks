```python
from mnspy import DerivadaDiscreta
import numpy as np
from tabulate import tabulate
```

# Diferenciación Numérica de Datos Discretos
A menudo, en aplicaciones de ingeniería y ciencia, no se tiene una función analítica, sino un conjunto de datos discretos obtenidos de un experimento o una simulación. Aún es posible estimar la derivada en estos puntos utilizando fórmulas de diferencias finitas.
 
**Problema:**

La siguiente tabla contiene mediciones de la posición ($x$) de una partícula en diferentes instantes de tiempo ($t$). Estimar la velocidad y la aceleración de la partícula en función del tiempo.

- **Velocidad ($v$)**: Es la primera derivada de la posición con respecto al tiempo ($v = dx/dt$).
- **Aceleración ($a$)**: Es la primera derivada de la velocidad con respecto al tiempo ($a = dv/dt$), o la segunda derivada de la posición ($a = d^2x/dt^2$).


# --- 1. Definición de los Datos ---

```python
# Vectores con los datos de tiempo (t) y posición (x).
t = np.array([4.0, 4.2, 4.4, 4.6,4.8, 5.0, 5.2, 5.4, 5.6, 5.8, 6.0, 6.2, 6.4, 6.6, 6.8, 7.0, 7.2, 7.4,7.6, 7.8, 8.0])
x = np.array([-5.87, -4.23, -2.55, -0.89, 0.67, 2.09, 3.31, 4.31, 5.06, 5.55, 5.78, 5.77, 5.52, 5.08, 4.46, 3.72, 2.88, 2.00, 1.10, 0.23, -0.59])
```

# --- 2. Cálculo de la Velocidad (Primera Derivada) ---

```python
# Usamos la clase `DerivadaDiscreta` para calcular la velocidad (v) a partir de los datos de tiempo (t) y posición (x).
# El modo 'centrada' utiliza diferencias finitas centradas para los puntos interiores, lo que generalmente da una mejor aproximación.
# Para los puntos extremos, la clase probablemente usa diferencias hacia adelante o hacia atrás.
vel = DerivadaDiscreta(t,x, modo='centrada')
```

```python
# Mostramos los resultados en una tabla. La columna 'v' contiene la velocidad calculada en cada instante 't'.
tabulate({'t':t, 'v':vel.derivada}, headers='keys', tablefmt='html')
```

```python
# Graficamos la posición (x) y la velocidad calculada (v) en función del tiempo (t).
vel.graficar('Derivada x contra t','$t$','$x$','$v$')
```

# --- 3. Cálculo de la Aceleración (Segunda Derivada) ---

```python
# Para hallar la aceleración (a), calculamos la derivada de la velocidad (vel.derivada) que acabamos de obtener.
ac = DerivadaDiscreta(t,vel.derivada, modo='centrada')
```

```python
# Mostramos los resultados de la aceleración en una tabla.
tabulate({'t':t, 'a':ac.derivada}, headers='keys', tablefmt='html')
```

```python
# Graficamos la velocidad (v) y la aceleración calculada (a) en función del tiempo (t).
ac.graficar('Derivada v contra t','$t$','$v$','$a$')
```
