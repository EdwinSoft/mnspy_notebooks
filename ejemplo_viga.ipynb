```python
from mnspy import Nodo, Viga, Ensamble
```

```python
n_1 = Nodo('1', 0, grados_libertad={'y': False, 'eje_z': False})
n_2 = Nodo('2', 2, grados_libertad={'y': True, 'eje_z': True})
n_3 = Nodo('3', 4, grados_libertad={'y': False, 'eje_z': True})
n_4 = Nodo('4', 10, grados_libertad={'y': False, 'eje_z': True})
n_5 = Nodo('5', 14, grados_libertad={'y': False, 'eje_z': False})
```

```python
n_2.agregar_fuerza_externa(-10,'y')
n_3.agregar_fuerza_externa(-10,'y')
n_4.agregar_fuerza_externa(-20,'eje_z')
```

```python
e_1 = Viga('1', n_1, n_2, E=2E7, I=1E-3)
e_2 = Viga('2', n_2, n_3, E=2E7, I=1E-3)
e_3 = Viga('3', n_3, n_4, E=2E7, I=1E-3)
e_4 = Viga('4', n_4, n_5, E=2E7, I=1E-3)
```

```python
e_3.agregar_carga_distribuida(-2)
e_4.agregar_carga_distribuida(-2)
```
