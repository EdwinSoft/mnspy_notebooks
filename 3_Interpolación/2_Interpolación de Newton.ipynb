# Interpolación Polinomial de Newton
El **polinomio interpolante de Newton en diferencias divididas** es una de las formas más populares y eficientes de realizar una interpolación polinomial. En lugar de resolver un sistema de ecuaciones lineales para encontrar los coeficientes, utiliza un cálculo recursivo llamado **diferencias divididas**.

La forma general del polinomio de grado $n$ es:
$$f_n(x) = b_0 + b_1(x-x_0) + \dots + b_n(x-x_0)(x-x_1)\dots(x-x_{n-1})$$
donde los coeficientes $b_i$ son las diferencias divididas finitas, que se calculan de forma sistemática en una tabla.

Una gran ventaja de esta forma es que si se añade un nuevo punto de datos, se pueden calcular los nuevos coeficientes sin tener que recalcular los antiguos.
 

## Ejemplo: Interpolación de $\ln(x)$
 
**Problema:**

Dados los siguientes puntos de la función $y = \ln(x)$, encontrar el polinomio interpolante de Newton y usarlo para estimar el valor de $\ln(2)$.

Puntos: $(1, \ln(1)), (4, \ln(4)), (6, \ln(6)), (5, \ln(5))$

```python
from mnspy import InterpolacionNewton
import numpy as np
```

```python
# Primero, definimos los vectores `x` e `y` con los puntos de datos.
# Nótese que para este método, los puntos no necesitan estar ordenados.
x = np.array([1., 4., 6., 5.])
y = np.log(x)
```

```python
# A continuación, creamos una instancia de la clase `InterpolacionNewton`, pasándole los puntos de datos.
inter = InterpolacionNewton(x, y)
```

```python
# Ahora, usamos el método `evaluar()` para interpolar el valor de `y` en `x=2`.
print(f"El valor interpolado en x=2 es: {inter.evaluar(2)}")
```

```python
# Podemos visualizar el resultado. La gráfica muestra los puntos originales,
# el polinomio de Newton que pasa por ellos y el punto interpolado (en rojo).
inter.graficar(2)
```

```python
# El método `obtener_polinomio()` nos muestra el polinomio en su forma de Newton,
# que es útil para entender la construcción del mismo.
inter.obtener_polinomio()
```

```python
# Si pasamos el argumento `expandido=True`, la biblioteca simplifica la expresión anterior
# para mostrar el polinomio en su forma canónica (potencias de x).
inter.obtener_polinomio(expandido=True)
```

```python
# Finalmente, podemos ver la tabla de diferencias divididas.
# La primera fila de esta tabla contiene los coeficientes b_0, b_1, b_2, ... del polinomio de Newton.
inter.mostrar_diferencias_divididas()
```
