## 1. Extracción de Información de una Tabla Discreta

**Problema:**

La tabla a continuación presenta un conjunto de propiedades para varios materiales granulares. Desarrolle una función en **Python** que tenga como argumentos el nombre del material y el nombre de la propiedad y devuelva el valor apropiado. Pruebe el programa con el material **Maíz** y la propiedad **Diámetro de partícula**.


| **Material**     | **Densidad absoluta** $[kg/m^{3}]$ | **Densidad a granel** $[kg/m^{3}]$ | **Porcentaje de vacío** | **Diámetro de partícula** $[mm]$ | **Factor de forma** |
|------------------|------------------------------------|------------------------------------|-------------------------|----------------------------------|---------------------|
| **Trigo**        | 1400                               | 865                                | 39.2                    | 3.61                             | 1.07                |
| **Arroz**        | 1457                               | 905                                | 37.9                    | 2.72                             | 1.04                |
| **Mijo**         | 1180                               | 727                                | 38.4                    | 1.99                             | 1.07                |
| **Polietileno**  | 922                                | 592                                | 35.8                    | 3.43                             | 1.02                |
| **Maíz**         | 1342                               | 743                                | 44.6                    | 7.26                             | 1.50                |
| **Poliestireno** | 1058                               | 641                                | 39.4                    | 1.56                             | 1.14                |
| **Cebada**       | 1279                               | 725                                | 43.4                    | 3.70                             | 1.14                |
| **Linaza**       | 1129                               | 703                                | 37.8                    | 2.09                             | 1.05                |


### Solución
Primero, almacenamos los datos de la tabla en estructuras de Python. Usaremos listas para las cabeceras (propiedades), los índices (materiales) y una lista de listas para los datos numéricos.


```python
propiedades=['Densidad absoluta', 'Densidad a granel', 'Porcentaje de vacío', 'Diámetro de partícula', 'Factor de forma']
material=['Trigo', 'Arroz', 'Mijo', 'Polietileno', 'Maíz', 'Poliestireno', 'Cebada', 'Linaza']
datos = [[1400,	865,	39.2,	3.61,	1.07],
	[1457,	905,	37.9,	2.72,	1.04],
	[1180,	727,	38.4,	1.99,	1.07],
	[922,	592,	35.8,	3.43,	1.02],
	[1342,	743,	44.6,	7.26,	1.50],
	[1058,	641,	39.4,	1.56,	1.14],
	[1279,	725,	43.4,	3.70,	1.14],
	[1129,	703,	37.8,	2.09,	1.05]]
```

```python
# Para visualizar los datos de forma clara, usamos la biblioteca `tabulate`.
from tabulate import tabulate
tabulate(datos, headers=propiedades, showindex=material, tablefmt='html')
```

```python
# Ahora, creamos la función `buscar`. Esta función encuentra el índice de la fila (material) y la columna (propiedad)
# y devuelve el valor correspondiente de la matriz de datos.
def buscar(mat: str, prop: str)->float:
    i = material.index(mat)
    j = propiedades.index(prop)
    return datos[i][j]
```

```python
# Probamos la función para obtener el 'Diámetro de partícula' del 'Maíz'.
buscar('Maíz', 'Diámetro de partícula')
```

 ## 2. Interpolación de Datos en una Tabla
 
 **Problema:**
 
La tabla presenta las propiedades de densidad y viscosidad del **aire** en función de la temperatura. Realice un programa en **Python** en el que se ingrese el valor de la temperatura y el nombre de la propiedad que se desea interpolar. Haga uso de **Numpy** y el método de **spline lineal** y pruébelo para hallar la *viscosidad* del aire para una $T$ de $210^\circ C$.


| $T\ [^\circ C]$ | $\rho\ [kg/m^{3}]$ | $\mu\ [ (Pa\cdot s)\times 10^{5} ]$ |
 |:---:|:---:|:---:|
 | -40 | 1.52 | 1.51 |
 | 0 | 1.29 | 1.71 |
 | 20 | 1.20 | 1.80 |
 | 50 | 1.09 | 1.95 |
 | 100 | 0.95 | 2.17 |
 | 150 | 0.84 | 2.38 |
 | 200 | 0.75 | 2.57 |
 | 250 | 0.68 | 2.75 |
 | 300 | 0.62 | 2.93 |
 | 400 | 0.53 | 3.25 |
 | 500 | 0.46 | 3.55 |

### Solución
Almacenamos los datos en arreglos de NumPy, que es ideal para operaciones numéricas.

```python
import numpy as np
from mnspy import SplineLineal
T= np.array([-40, 0, 20, 50, 100, 150, 200, 250, 300, 400, 500])
den= np.array([1.52, 1.29, 1.20, 1.09, 0.95, 0.84, 0.75, 0.68, 0.62, 0.53, 0.46])
vis= np.array([1.51, 1.71, 1.80, 1.95, 2.17, 2.38, 2.57, 2.75, 2.93, 3.25, 3.55])
```

```python
# Visualizamos los datos en una tabla.
tabulate({'T': T, r'$\rho$': den, r'$\mu$': vis}, headers='keys', tablefmt='html')
```

```python
# Creamos una función `inter` que toma la temperatura y el nombre de la propiedad.
# Un diccionario ayuda a seleccionar el conjunto de datos correcto (densidad o viscosidad).
# Luego, se instancia la clase `SplineLineal` de `mnspy` para realizar la interpolación.
def inter(tem: float, prop:str):
    dic ={'densidad': den, 'viscosidad': vis}
    if prop not in dic.keys():
        print('Propiedad no valida')
        return
    tabla = SplineLineal(T, dic[prop])
    return tabla.evaluar(tem)
```

```python
# Probamos la función para interpolar la viscosidad a 210°C.
inter(210,'viscosidad')
```

 ## 3. Interpolación de Datos desde un Archivo de Excel
 
 **Problema:**
 
Importe el archivo de Excel **agua.xlsx** usando el paquete **pandas**. El archivo contiene propiedades del agua saturada. Realice un programa en **Python** en el que se ingrese la temperatura y el nombre de la propiedad a interpolar. Use el método de **spline lineal** y pruébelo para hallar la entalpía de líquido saturado (*hf*) del agua para una $T$ de $157^\circ C$.

### Solución
Usamos la biblioteca `pandas` para leer el archivo de Excel y cargarlo en un DataFrame, que es una estructura de datos tabular muy potente. El argumento `skiprows=1` se usa para omitir la primera fila del archivo si contiene información no deseada.


```python
import pandas as pd
```

```python
df = pd.read_excel('agua.xlsx', skiprows=1)
df
```

```python
from mnspy import SplineLineal
def inter(tem: float, prop:str):
    if prop not in df.keys():
        print('Propiedad no valida')
        return
    tabla = SplineLineal(df['T, °C'], df[prop])
    return tabla.evaluar(tem)
```

```python
# Probamos la función para encontrar la entalpía 'hf, kJ/kg' a 157°C.
inter(157,'hf')
```

```python
inter(275,'P, MPa')
```
