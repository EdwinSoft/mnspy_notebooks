# Interpolación Polinomial: Formas de Lagrange y Newton

 La **interpolación polinomial** es un método para encontrar un polinomio que pasa exactamente por un conjunto de puntos de datos. Este notebook compara dos de las formas más comunes de construir este polinomio: la forma de **Lagrange** y la forma de **Newton**.

 Aunque las fórmulas se ven diferentes, ambas producen el mismo y único polinomio de interpolación para un conjunto de datos dado.

 ---
 ## Problema

 Dados los siguientes puntos de datos:

 |   x |   y |
 |----:|----:|
 | 1   | 2   |
 | 2.2 | 2.8 |
 | 3.4 | 3   |
 | 4.8 | 3.2 |
 | 6   | 4   |
 | 7   | 5   |

 **(a)** Escribir el polinomio en **forma de Lagrange** que pasa por los puntos y usarlo para calcular el valor interpolado de $y$ en $x = 5.4$.

 **(b)** Escribir el polinomio en **forma de Newton** que pasa por los puntos y usarlo para calcular el valor interpolado de $y$ en $x = 5.4$.


```python
from mnspy import InterpolacionLagrange, InterpolacionNewton
import numpy as np
```

### Parte (a): Polinomio de Interpolación de Lagrange
 #### Paso 1: Definición de los datos

```python
# Definimos los vectores con los puntos de datos x e y.
x = np.array([1, 2.2, 3.4, 4.8, 6, 7])
y = np.array([2, 2.8, 3.0, 3.2, 4, 5])
```

```python
# Mostramos los datos en una tabla para mayor claridad.
from tabulate import tabulate
tabulate({'x': x, 'y': y},headers='keys',tablefmt='html')
```

#### Paso 2: Creación y evaluación del polinomio
Creamos una instancia de la clase `InterpolacionLagrange` con los datos.

```python
inter = InterpolacionLagrange(x,y)
```

El método `evaluar()` calcula el valor del polinomio en el punto x=5.4.

```python
inter.evaluar(5.4)
```

#### Paso 3: Visualización

```python
inter.graficar(5.4)
```

#### Paso 4: Obtención de la forma del polinomio

```python
# El método `obtener_polinomio()` muestra el polinomio en su forma de Lagrange,
# que es una suma ponderada de los polinomios base de Lagrange.
inter.obtener_polinomio()
```

```python
# Al expandir y simplificar la forma de Lagrange, obtenemos el polinomio en su forma canónica.
inter.obtener_polinomio(expandido=True)
```

### Parte (b): Polinomio de Interpolación de Newton
#### Paso 1: Creación y evaluación del polinomio
Creamos una instancia de la clase `InterpolacionNewton`.

```python
inter = InterpolacionNewton(x,y)
```

```python
# Evaluamos en el mismo punto x=5.4. El resultado debe ser idéntico al del método de Lagrange.
inter.evaluar(5.4)
```

#### Paso 2: Visualización

```python
# La gráfica generada es la misma, ya que el polinomio subyacente es el mismo.
inter.graficar(5.4)
```

#### Paso 3: Obtención de la forma del polinomio
El método `obtener_polinomio()` muestra el polinomio en la forma de Newton, que se construye utilizando las diferencias divididas.

```python
inter.obtener_polinomio()
```

```python
# Al expandir la forma de Newton, obtenemos el mismo polinomio canónico que con Lagrange.
inter.obtener_polinomio(expandido=True)
```

#### Paso 4: Tabla de Diferencias Divididas
El método de Newton se basa en la tabla de diferencias divididas.
El método `mostrar_diferencias_divididas()` la calcula y la presenta los coeficientes del polinomio de Newton son los valores de la primera fila de esta tabla.

```python
inter.mostrar_diferencias_divididas()
```
