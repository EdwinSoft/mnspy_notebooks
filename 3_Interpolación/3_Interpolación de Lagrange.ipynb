# Interpolación Polinomial de Lagrange
La **interpolación de Lagrange** es otro método para encontrar el polinomio único que pasa por un conjunto de puntos. A diferencia del método de Newton, que utiliza diferencias divididas, Lagrange construye el polinomio como una suma ponderada de los valores $y_i$.
 
 La fórmula del polinomio de Lagrange es:
 $$ P(x) = \sum_{i=0}^{n} y_i L_i(x) $$
 
 donde $L_i(x)$ son los **polinomios base de Lagrange**, definidos como:
 $$ L_i(x) = \prod_{j=0, j\neq i}^{n} \frac{x-x_j}{x_i-x_j} $$
 
 Cada polinomio base $L_i(x)$ tiene la propiedad de que vale 1 en $x_i$ y 0 en todos los demás puntos $x_j$. Esto asegura que el polinomio final $P(x)$ pase exactamente por todos los puntos de datos.
 
 Aunque conceptualmente es elegante, puede ser computacionalmente menos eficiente que el método de Newton si se necesita evaluar el polinomio en muchos puntos o si se añaden nuevos datos con frecuencia.


## Ejemplo: Interpolación de $\ln(x)$
 
**Problema:**

Dados los siguientes puntos de la función $y = \ln(x)$, encontrar el polinomio interpolante de Lagrange y usarlo para estimar el valor de $\ln(2)$.

Puntos: $(1, \ln(1)), (4, \ln(4)), (6, \ln(6)), (5, \ln(5))$

```python
from mnspy import InterpolacionLagrange
import numpy as np
```

```python
# Primero, definimos los vectores `x` e `y` con los puntos de datos.
# Al igual que en el método de Newton, los puntos no necesitan estar ordenados.
x = np.array([1., 4., 6., 5.])
y = np.log(x)
```

```python
# A continuación, creamos una instancia de la clase `InterpolacionLagrange`, pasándole los puntos de datos.
inter = InterpolacionLagrange(x, y)
```

```python
# Ahora, usamos el método `evaluar()` para interpolar el valor de `y` en `x=2`.
print(f"El valor interpolado en x=2 es: {inter.evaluar(2)}")
```

```python
# Podemos visualizar el resultado. La gráfica muestra los puntos originales,
# el polinomio de Lagrange que pasa por ellos y el punto interpolado (en rojo).
inter.graficar(2)
```

```python
# El método `obtener_polinomio()` nos muestra el polinomio en su forma de Lagrange,
# que es una suma de los polinomios base ponderados por los valores de y.
inter.obtener_polinomio()
```

```python
# Si pasamos el argumento `expandido=True`, la biblioteca simplifica la expresión anterior
# para mostrar el polinomio en su forma canónica (potencias de x).
# Este resultado debe ser idéntico al obtenido con el método de Newton, ya que el polinomio interpolante es único.
inter.obtener_polinomio(expandido=True)
```
