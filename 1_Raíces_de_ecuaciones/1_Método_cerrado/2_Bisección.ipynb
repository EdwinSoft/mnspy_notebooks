# Método de la Bisección
El **método de la bisección** es un algoritmo de búsqueda de raíces que divide repetidamente un intervalo por la mitad y luego selecciona el subintervalo en el que debe existir una raíz para el siguiente procesamiento. Es un método muy simple y robusto, pero relativamente lento.
 
Para que el método funcione, se debe partir de un intervalo `[a, b]` donde la función `f(x)` sea continua y `f(a)` y `f(b)` tengan signos opuestos. Esto, según el Teorema del Valor Intermedio, garantiza que existe al menos una raíz en el intervalo.

## Ejemplo 1: Función Polinómica

```python
from mnspy import Biseccion
import numpy as np
# se habilita los gráficos interactivos de la biblioteca Matplotlib para habilitar el zoom.
%matplotlib widget
```

```python
# Definimos la función a la que le queremos encontrar la raíz:
def f(x):
    return x ** 3 - 10 * x ** 2 + 5
```

```python
# Instanciamos la clase Biseccion con la función, el intervalo [0, 1] y una tolerancia de 0.01.
bis = Biseccion(f, 0, 1, tol=1e-2)
# Generamos la tabla de iteraciones para ver el proceso paso a paso:
bis.generar_tabla()
```

```python
# Graficamos el proceso de convergencia:
bis.graficar()
```

```python
# Finalmente, obtenemos la solución encontrada:
bis.solucion()
```

## Ejemplo 5.3 (Chapra, 5ª ed.)

**Problema:**

Utilice el método de la bisección para determinar el coeficiente de arrastre $c$ necesario para que un paracaidista de masa $m = 68.1\,kg$ tenga una velocidad de $40\,m/s$ después de una caída libre de $t = 10\,s$. Nota: La aceleración de la gravedad es $9.8\,m/s^{2}$. En el rango de 12 a 16 y con un error relativo aproximado $\varepsilon_{a}=0.5\%$.
$$f(c)=\dfrac{gm}{c}\left(1-e^{-(c/m)t}\right)-v$$

$$f(c)=\dfrac{9.8\times 68.1}{c}\left(1-e^{-(c/68.1)\times 10}\right)-40$$
$$[12,16]; \varepsilon_{s}=0.5\%$$

Sustituyendo los valores dados:
$$f(c)=\dfrac{9.8\times 68.1}{c}\left(1-e^{-(c/68.1)\times 10}\right)-40$$
El intervalo inicial es $[12, 16]$ y la tolerancia porcentual es $\varepsilon_{s}=0.5\%$.

```python
# Definimos la función del problema del paracaidista:
def f(c):
    m = 68.1
    v = 40
    t = 10
    g = 9.8
    return (g*m/c)*(1-np.exp(-c*t/m))-v
```

```python
# Instanciamos Biseccion con el intervalo [12, 16] y una tolerancia del 0.5%.
bis = Biseccion(f,12,16,0.5)
# Tabla de iteraciones:
bis.generar_tabla()
```

```python
# Gráfica del proceso:
bis.graficar()
```

```python
# Solución encontrada:
bis.solucion()
```

## Ejemplo 1 (Burden, Análisis numérico 10ª ed.)

Show that $f(x)=x^3 + 4x^2 - 10 = 0$ has a root in [1, 2], and use the Bisection method to determine an approximation to the root that is accurate to at least within $10^{-4}$.

```python
# Definimos la función del problema:
def f(x):
    return x**3+4*x**2-10
```

```python
# Instanciamos Biseccion. El parámetro tipo_error='n' indica que la tolerancia es de 10 elevado a la -4.
bis = Biseccion(f,1,2,4, tipo_error='n')
# Tabla de iteraciones:
bis.generar_tabla()
```

```python
# Gráfica del proceso (con más puntos para suavizar la curva):
bis.graficar(n_puntos=1000)
```
