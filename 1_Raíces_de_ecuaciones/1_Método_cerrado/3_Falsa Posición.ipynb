# Método de la Falsa Posición
El **método de la falsa posición** (también conocido como *regula falsi*) es una alternativa al método de la bisección. Al igual que la bisección, comienza con un intervalo `[a, b]` que contiene una raíz, es decir, donde `f(a)` y `f(b)` tienen signos opuestos.
 
 La principal diferencia es que, en lugar de dividir el intervalo por la mitad, este método traza una línea recta entre los puntos `(a, f(a))` y `(b, f(b))`. La nueva aproximación de la raíz es el punto donde esta línea recta cruza el eje x.
 
 Este enfoque generalmente converge más rápido que la bisección, aunque en algunos casos puede estancarse si uno de los extremos del intervalo permanece fijo durante muchas iteraciones.
 

## Ejemplo 1: Función Polinómica

```python
from mnspy import FalsaPosicion
# se habilita los gráficos interactivos de la biblioteca Matplotlib para habilitar el zoom.
%matplotlib widget
```

```python
# Definimos la función a la que le queremos encontrar la raíz:
def f(x):
    return x ** 3 - 10 * x ** 2 + 5
```

```python
# Instanciamos la clase FalsaPosicion con la función, el intervalo [0, 1] y una tolerancia de 0.01. 
fp = FalsaPosicion(f, 0, 1, tol=1e-2)
```

```python
# Generamos la tabla de iteraciones para ver el proceso paso a paso:
fp.generar_tabla()
```

```python
# Graficamos el proceso de convergencia:
fp.graficar()
```

```python
# Finalmente, obtenemos la solución encontrada:
fp.solucion()
```
