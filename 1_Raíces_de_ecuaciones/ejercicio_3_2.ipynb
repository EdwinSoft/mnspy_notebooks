# Ejercicio 3.2: Comparación de Métodos de Búsqueda de Raíces
  
 **Problema:**
  
 Basado en el ejercicio 3.2 del libro *Numerical Methods for Engineers and Scientists* de Amos Gilat y Vish Subramaniam.
  
 Determinar la raíz de la función:
 $$f(x) = x – 2e^{-x}$$
  
 Utilizando los siguientes métodos y condiciones iniciales:
  
 **(a)** **Método de Bisección:** Empezar con el intervalo $[a, b] = [0, 1]$ y realizar las primeras tres iteraciones.
  
 **(b)** **Método de la Secante:** Empezar con los puntos $x_1 = 0$ y $x_2 = 1$, y realizar las primeras tres iteraciones.
  
 **(c)** **Método de Newton-Raphson:** Empezar con el punto inicial $x_1 = 1$ y realizar las primeras tres iteraciones.

```python
import numpy as np
from mnspy import Biseccion, Secante, NewtonRaphson
```

## --- Definición de la Función ---

```python
# Primero, definimos la función f(x) para la cual buscaremos la raíz.
def f(x):
    return x-2*np.exp(-x)
```

### --- (a) Método de Bisección ---
Creamos una instancia de la clase `Biseccion` con la función, el intervalo [0, 1] y un máximo de 3 iteraciones.

```python
bis = Biseccion(f,0,1,max_iter=3)
```

El método `graficar()` muestra la función y cómo el intervalo se va reduciendo en cada iteración.

```python
bis.graficar()
```

El método `generar_tabla()` muestra los resultados numéricos de las 3 iteraciones.

```python
bis.generar_tabla()
```

### --- (b) Método de la Secante ---
Creamos una instancia de la clase `Secante` con los puntos iniciales 0 y 1, y un máximo de 3 iteraciones.

```python
sec = Secante(f,0,1, max_iter=3)
# La gráfica muestra cómo se trazan las líneas secantes para aproximar la raíz.
sec.graficar()
```

```python
# La tabla muestra el progreso de las aproximaciones en cada paso.
sec.generar_tabla()
```

### --- (c) Método de Newton-Raphson ---
Para este método, necesitamos definir también la primera derivada de la función, df(x).

```python
def df(x):
    return 1+2*np.exp(-x)
```

Creamos la instancia de `NewtonRaphson` con f(x), df(x), el punto inicial 1 y un máximo de 3 iteraciones.

```python
nr = NewtonRaphson(f,df,1, max_iter=3)
# La gráfica muestra cómo se trazan las líneas tangentes para encontrar la raíz.
nr.graficar()
```

```python
# La tabla muestra la rápida convergencia del método.
nr.generar_tabla()
```
