# Método de Brent
El **método de Brent** es un algoritmo de búsqueda de raíces que combina la seguridad del método de la bisección con la velocidad de convergencia de métodos abiertos como el de la secante y la interpolación cuadrática inversa.
 
 Es considerado uno de los métodos más robustos y eficientes para encontrar raíces de funciones de una variable, ya que:
 1.  Garantiza la convergencia siempre que se parta de un intervalo `[a, b]` donde la raíz esté contenida (`f(a)` y `f(b)` tienen signos opuestos).
 2.  Intenta usar los métodos más rápidos (secante o interpolación cuadrática) siempre que sea posible.
 3.  Recurre al método de la bisección si los métodos rápidos no están convergiendo adecuadamente, evitando así la divergencia.
 
 Debido a su fiabilidad y eficiencia, es el método de elección en muchas bibliotecas científicas.


## Ejemplo: Función Polinómica
**Problema:**

Encontrar la raíz de la función:
 $$f(x) = \dfrac{x^3 + 7x^2 - 40x - 100}{50}$$
 en el intervalo `[0, 8]` con una tolerancia de `0.0001`.

```python
from mnspy import Brent
# se habilita los gráficos interactivos de la biblioteca Matplotlib para habilitar el zoom.
%matplotlib widget
```

```python
# Primero, definimos la función f(x):
def f(x):
    return (x ** 3 + 7 * x ** 2 - 40 * x - 100)/50
```

```python
# Instanciamos la clase `Brent`, proporcionando la función, el intervalo inicial [a, b] y la tolerancia deseada.
br = Brent(f, 0.0, 8.0, 0.0001)
```

```python
# Personalizamos el formato de la tabla y la generamos para observar el proceso de convergencia.
br.formato_tabla('E_a','0.8%')
br.generar_tabla()
```

```python
# Graficamos la función y el proceso de búsqueda de la raíz.
br.graficar()
```

```python
# Finalmente, obtenemos el valor de la raíz encontrada.
br.solucion()
```
