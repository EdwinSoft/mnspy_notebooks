# Método de la Wegstein
El **método de Wegstein** es una técnica de aceleración para la iteración de punto fijo. Es particularmente útil para problemas donde la iteración de punto fijo converge lentamente o diverge.
 
 El método modifica la siguiente estimación de la raíz basándose en la pendiente de la línea que une los dos últimos puntos. La fórmula de recurrencia es:
 
 $$x_{i+1} = q x_i + (1-q) g(x_i)$$
 
 donde $q$ es un factor de aceleración que se calcula en cada paso:
 
 $$q = \dfrac{s}{s-1} \quad \text{con} \quad s = \dfrac{g(x_i) - g(x_{i-1})}{x_i - x_{i-1}}$$
 
 Al igual que el método de la secante, requiere dos puntos iniciales para comenzar.


```python
from mnspy import Wegstein
import numpy as np
# se habilita los gráficos interactivos de la biblioteca Matplotlib para habilitar el zoom.
%matplotlib widget
```

 ## Ejemplo 1
 **Problema:**
 
 Para la ecuación $x=e^{-x}$ halle la por el método de Wegstein con $x_{0}=0$ y $x_{1}=0.25$ con un $\varepsilon_{a}$ inferior a $1\times10^{-3}\%$

```python
# Definimos la función g(x) para la iteración:
def f(x):
    return np.exp(-x)
```

```python
# Instanciamos la clase Wegstein con los dos puntos iniciales y la tolerancia.
we = Wegstein(f, 0, 0.25, 1E-3)
```

```python
# Personalizamos el formato de la tabla y generamos las iteraciones.
we.formato_tabla('E_a','0.6%')
we.formato_tabla('f','0.10f')
# La tabla de iteraciones muestra el proceso paso a paso (se ingresa como parámetro el valor de la solución para que calcule el error teórico):
we.generar_tabla(0.567143140453502)
```

```python
# Graficamos el proceso de convergencia. La solución es la intersección entre y=x y y=g(x). 
we.graficar()
```

```python
# Finalmente, obtenemos la solución.
we.solucion()
```
