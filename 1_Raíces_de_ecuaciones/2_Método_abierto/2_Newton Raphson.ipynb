# Método de Newton Raphson
El **método de Newton-Raphson** es uno de los algoritmos más utilizados y eficientes para encontrar raíces de ecuaciones. Es un método abierto que requiere un único punto de partida y el conocimiento de la primera derivada de la función.
 
 La idea fundamental es aproximar la función $f(x)$ por su línea tangente en el punto actual $x_i$. La siguiente aproximación de la raíz, $x_{i+1}$, se toma como el punto donde esta línea tangente cruza el eje x. Esto conduce a la fórmula de recurrencia:
 
 $$x_{i+1} = x_i - \dfrac{f(x_i)}{f'(x_i)}$$
 
 Este método generalmente converge muy rápidamente (convergencia cuadrática) si el punto inicial está lo suficientemente cerca de la raíz. Sin embargo, puede diverger si la derivada es cercana a cero o si el punto inicial está mal elegido.


```python
from mnspy import NewtonRaphson
# se habilita los gráficos interactivos de la biblioteca Matplotlib para habilitar el zoom.
%matplotlib widget
```

## Ejemplo 1: Función Polinómica
 **Problema:**

Calcular la raíz de la ecuación:
$$x^{4}-6.4x^{3}+6.45x^{2}+20.538x-31.752$$ por el método de Newton Raphson con un $x_{inicial}=2$ y una toleracia del $0.1\%$, además conociendo que su derivada es:
$$4x^{3}-19.2x^{2}+12.9x+20.538$$

Primero, definimos la función $f(x)$ y su derivada $f'(x)$.

```python
def f(x):
    return x ** 4 - 6.4 * x ** 3 + 6.45 * x ** 2 + 20.538 * x - 31.752

def df(x):
    return 4.0 * x ** 3 - 19.2 * x ** 2 + 12.9 * x + 20.538
```

```python
# "Ahora, creamos una instancia de la clase `NewtonRaphson`, proporcionando la función, su derivada, el punto inicial y la tolerancia.
nr = NewtonRaphson(f, df=df, x=2, tol=0.1)
# La gráfica visualiza la función y cómo las líneas tangentes se aproximan a la raíz en cada paso.
nr.graficar()
```

```python
# La tabla de iteraciones muestra cómo el método converge rápidamente a la solución.
nr.generar_tabla()
```

```python
# Finalmente, obtenemos el valor de la raíz encontrada.
nr.solucion()
```
