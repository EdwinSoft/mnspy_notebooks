# Método de la Secante Modificada
El **método de la secante modificada** es una variación del método de la secante que evita la necesidad de dos puntos iniciales. En su lugar, utiliza un único punto inicial $x_i$ y un pequeño valor de perturbación $\delta$.
 
 La derivada se aproxima de la siguiente manera:
 
 $$f'(x_i) \approx \dfrac{f(x_i + \delta x_i) - f(x_i)}{\delta x_i}$$
 
 Sustituyendo esta aproximación en la fórmula de Newton-Raphson, obtenemos la fórmula de recurrencia para la secante modificada:
 
 $$x_{i+1} = x_i - \dfrac{\delta x_i f(x_i)}{f(x_i + \delta x_i) - f(x_i)}$$
 
 Este método es útil cuando la derivada es difícil de calcular y se quiere evitar el uso de dos puntos iniciales. El valor de $\delta$ es crucial: si es muy pequeño, puede causar errores de redondeo; si es muy grande, puede llevar a una convergencia lenta o a la divergencia.

## Ejemplo: Función Polinómica
**Problema:**

Calcular la raíz de la ecuación:
$$x^{4}-6.4x^{3}+6.45x^{2}+20.538x-31.752$$ por el método de Secante Modificadacon un $x_{inicial}=0$ y una toleracia del $0.1\%$, 

```python
# Primero, definimos la función f(x):
def f(x):
    return x ** 4 - 6.4 * x ** 3 + 6.45 * x ** 2 + 20.538 * x - 31.752
```

Ahora, creamos una instancia de la clase `SecanteModificada`.

Proporcionamos la función, el punto inicial `x`, el valor de perturbación `delta` y la tolerancia. 

```python
sm = SecanteModificada(f, x=0, delta=0.1, tol=0.1)
# Graficamos el proceso para visualizar la convergencia.
sm.graficar()
```

```python
# La tabla de iteraciones muestra el proceso paso a paso:
sm.generar_tabla()
```

```python
# Finalmente, obtenemos el valor de la raíz encontrada:
sm.solucion()
```
