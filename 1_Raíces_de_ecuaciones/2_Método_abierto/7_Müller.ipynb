# Método de Müller
El **método de Müller** es una generalización del método de la secante. Mientras que el método de la secante utiliza una línea recta que pasa por dos puntos para aproximar la raíz, el método de Müller utiliza una **parábola** que pasa por **tres puntos**.
 
 La intersección de esta parábola con el eje x se toma como la siguiente aproximación de la raíz. Este enfoque de segundo orden generalmente converge más rápido que el método de la secante.
 
 Una ventaja significativa del método de Müller es que, al resolver una ecuación cuadrática en cada paso, puede **encontrar raíces complejas**, algo que los métodos anteriores no pueden hacer.
 

```python
from mnspy import Muller
import numpy as np
# se habilita los gráficos interactivos de la biblioteca Matplotlib para habilitar el zoom.
%matplotlib widget
```

## Ejemplo 1
**Problema:**

Para la ecuación $\sin(x)$ halle la por el método de Müller con $x_{0}=5$ con un $\varepsilon_{s}$ inferior a $1\times10^{-3}\%$

```python
# Primero, definimos la función f(x):
def f(x):
    return np.sin(-x)
```

```python
# Instanciamos la clase `Muller`, proporcionando la función, el punto inicial y la tolerancia.
mu = Muller(f, 5, tol=1E-3)
```

```python
# Personalizamos el formato de la tabla y la generamos para ver el proceso.
mu.formato_tabla('E_a','0.6%')
mu.generar_tabla()
```

```python
# Graficamos el proceso de convergencia.
mu.graficar()
```

```python
# Finalmente, obtenemos el valor de la raíz encontrada.
mu.solucion()
```
