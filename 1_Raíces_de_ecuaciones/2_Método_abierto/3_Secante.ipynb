# Método de la Secante
El **método de la secante** es una alternativa popular al método de Newton-Raphson. Es un método abierto que, al igual que Newton-Raphson, utiliza una aproximación lineal para encontrar la siguiente estimación de la raíz.
 
 La principal ventaja del método de la secante es que **no requiere el cálculo de la derivada** de la función. En su lugar, aproxima la derivada utilizando una diferencia finita hacia atrás, basándose en los dos últimos puntos de la iteración:
 
 $$f'(x_i) \approx \dfrac{f(x_{i-1})-f(x_i)}{x_{i-1}-x_i}$$
 
 Sustituyendo esta aproximación en la fórmula de Newton-Raphson, obtenemos la fórmula de recurrencia del método de la secante:
 
 $$x_{i+1} = x_i - \dfrac{f(x_i)(x_{i-1}-x_i)}{f(x_{i-1})-f(x_i)}$$
 
 Para comenzar, el método requiere dos puntos iniciales, $x_0$ y $x_1$, que no necesitan necesariamente encerrar la raíz.


```python
from mnspy import Secante
# se habilita los gráficos interactivos de la biblioteca Matplotlib para habilitar el zoom.
%matplotlib widget
```

## Ejemplo: Función Polinómica
 **Problema:**


Calcular la raíz de la ecuación:
$$x^{4}-6.4x^{3}+6.45x^{2}+20.538x-31.752$$ por el método de Secante iniciando en $x_{0}=0$ y $x_{1}=1$ y una toleracia del $0.1\%$, 

```python
#Primero, definimos la función f(x):
def f(x):
    return x ** 4 - 6.4 * x ** 3 + 6.45 * x ** 2 + 20.538 * x - 31.752
```

```python
# Ahora, creamos una instancia de la clase `Secante`, proporcionando la función, los dos puntos iniciales y la tolerancia.
sm = Secante(f, x_0=0,x_1=1, tol=0.1)
# Graficamos el proceso. La gráfica muestra cómo la línea secante (que pasa por los dos últimos puntos) se aproxima a la raíz en cada paso.
sm.graficar()
```

```python
# La tabla de iteraciones muestra la convergencia paso a paso:
sm.generar_tabla()
```

```python
# Finalmente, obtenemos el valor de la raíz encontrada:
sm.solucion()
```
