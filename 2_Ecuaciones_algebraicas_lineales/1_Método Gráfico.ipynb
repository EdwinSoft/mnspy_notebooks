# Método Gráfico para Sistemas de Ecuaciones Lineales
El **método gráfico** es una forma visual de resolver un sistema de ecuaciones lineales. Consiste en graficar cada ecuación como una línea recta en un plano cartesiano. La solución del sistema es el punto de intersección de todas las líneas.
 
 Este método es muy intuitivo pero generalmente solo es práctico para sistemas de dos ecuaciones con dos incógnitas (sistemas 2x2), ya que en dimensiones más altas la visualización se vuelve compleja o imposible.


```python
from mnspy import EcuacionesAlgebraicasLineales
import numpy as np
# se habilita los gráficos interactivos de la biblioteca Matplotlib para habilitar el zoom.
%matplotlib widget
```

## Ejemplo: Sistema 2x2
 
**Problema:**
 
Solucionar el siguiente sistema de ecuaciones gráficamente:

\begin{equation}
    3x_{0}+2x_{1}=18
\end{equation}
\begin{equation}
    -x_{0}+2x_{1}=2
\end{equation}

En forma matricial, el sistema se expresa como:
\begin{equation}
		\begin{split}
			\begin{bmatrix}
				3 & 2 \\
				-1 & 2        
			\end{bmatrix}\cdot\begin{bmatrix}
				x_{0} \\
				x_{1}    
			\end{bmatrix}=\begin{bmatrix}
				18 \\
				2       
			\end{bmatrix}
		\end{split}
	\end{equation}
 $$A\cdot x=b$$

 ## Solución con mnspy

```python
# Definimos la matriz de coeficientes A y el vector de términos independientes b:
A = np.array([[3,2],[-1,2]])
b = np.array([[18],[2]])
```

```python
# Creamos una instancia de la clase `EcuacionesAlgebraicasLineales`:
sol = EcuacionesAlgebraicasLineales(A, b)
# Mostramos el sistema de ecuaciones para verificar que se ha cargado correctamente.
sol.mostrar_sistema()
```

```python
# Graficamos el sistema. Cada línea representa una ecuación y el punto de intersección es la solución.
sol.graficar()
```

## Verificación
 
De la gráfica, se puede observar que las dos líneas se cruzan en el punto **(4, 3)**. Por lo tanto, la solución del sistema es $x_0 = 4$ y $x_1 = 3$.

Podemos verificar este resultado multiplicando la matriz `A` por el vector solución. El resultado debe ser igual al vector `b`.


```python
np.linalg.matmul(A,[[4.0],[3.0]])
```
